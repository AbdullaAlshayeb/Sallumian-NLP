# Stemming

Stemming is the process of reducing words to their root form by removing suffixes.

For example:

running → run  
playing → play  
studies → studi  

Unlike lemmatization, stemming does not use a dictionary.  
Instead, it applies simple rule-based transformations to cut word endings.

Stemming helps reduce the number of unique words in a dataset, which can improve the performance of many NLP models.

## Import Libraries

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

nltk.download("punkt")
```

## Example Text

```python
text = """
My name is Abdullah Hamdi. I am an AI engineer who is passionate about machine learning and natural language processing.
I am currently building this NLP repository to strengthen my understanding of text processing and machine learning techniques.
At the moment, I am working on this project while completing my military service in Salloum.
Even though the environment is challenging and resources are limited, I believe that continuous learning is possible anywhere.
My goal with this repository is to practice, experiment, and improve my skills step by step so that I can become a better AI engineer in the future.
"""
```

## Tokenize the Text

```python
tokens = word_tokenize(text)

print(tokens[:20])
```

## Initialize the Stemmer

```python
stemmer = PorterStemmer()
```

## Apply Stemming

```python
stemmed_words = [stemmer.stem(word) for word in tokens]

print(stemmed_words[:20])
```

## Compare Original vs Stemmed Text

```python
for word in tokens[:20]:
    print(word, "→", stemmer.stem(word))
```

## Limitations of Stemming

Stemming can produce non-real words.

Example:

studies → studi
engineer → engin

Because stemming cuts suffixes blindly.

This is why lemmatization (next notebook) is usually preferred.

## Key Takeaways

- Stemming reduces words to their root form.
- It removes suffixes using rule-based algorithms.
- It helps reduce vocabulary size in NLP tasks.
- Stemming can produce non-real words because it does not use dictionaries.

# Text Analysis

Text analysis is the process of exploring and understanding text data
before applying machine learning models.

This step helps us understand:

- most frequent words
- sentence length
- word distribution
- dataset characteristics

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

## Example Dataset

```python
texts = [
    "Natural language processing is amazing",
    "Machine learning is powerful",
    "Artificial intelligence is the future",
    "Deep learning improves many applications",
    "Natural language processing enables computers to understand text"
]

df = pd.DataFrame({"text": texts})
df
```

## Text Length Analysis

```python
df["word_count"] = df["text"].apply(lambda x: len(x.split()))

df
```

## Plot Word Count Distribution

```python
plt.hist(df["word_count"])
plt.title("Word Count Distribution")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
plt.show()

# This helps identify:
#   -short texts
#   -long texts
#   -dataset imbalance
```

## Most Frequent Words

```python
# First Combine All Text
all_words = " ".join(df["text"]).lower().split()

# Count Frequencies 
word_freq = Counter(all_words)

word_freq.most_common(10)
```

## Plot Most Frequent Words 

```python
words, counts = zip(*word_freq.most_common(10))

plt.bar(words, counts)
plt.title("Most Frequent Words")
plt.xticks(rotation=45)
plt.show()
```

## Key Takeaways

Text analysis helps us understand:

- dataset structure
- vocabulary size
- common words
- text length distribution

These insights guide preprocessing and feature engineering steps.

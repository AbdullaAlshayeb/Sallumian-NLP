# Tokenization

Tokenization is the process of breaking text into smaller units called tokens.

Tokens can represent:
- sentences
- words
- characters

Tokenization is an essential preprocessing step in Natural Language Processing because machine learning models cannot directly process raw text.

In this notebook we will explore:
- Sentence Tokenization
- Word Tokenization
- Tokenization using Python
- Tokenization using NLP libraries

## Import Libraries

```python
import nltk
import spacy

nltk.download("punkt")
nlp = spacy.load("en_core_web_sm")
```

## Example Dataset

```python
text = """
My name is Abdullah Hamdi. I am an AI engineer who is passionate about machine learning and natural language processing. 
I am currently building this NLP repository to strengthen my understanding of text processing and machine learning techniques. 
At the moment, I am working on this project while completing my military service in Salloum. 
Even though the environment is challenging and resources are limited, I believe that continuous learning is possible anywhere. 
My goal with this repository is to practice, experiment, and improve my skills step by step so that I can become a better AI engineer in the future.
"""
```

## Sentence Tokenization

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)

print(sentences)
```

## Word Tokenization

```python
from nltk.tokenize import word_tokenize

words = word_tokenize(text)

print(words)
```

## Tokenization using spaCy

```python
doc = nlp(text)

tokens = [token.text for token in doc]

print(tokens)
```

## Sentence Tokenization using spaCy

```python
sentences_spacy = [sent.text for sent in doc.sents]

print(sentences_spacy)
```

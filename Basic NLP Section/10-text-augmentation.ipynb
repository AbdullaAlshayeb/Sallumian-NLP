# Text Augmentation

Text augmentation is the process of generating new text samples from existing data.

It is commonly used to increase dataset size and improve model generalization.

## Import Library

```python
!pip install nlpaug
```

```python
import nlpaug.augmenter.word as naw
import nltk

nltk.download('averaged_perceptron_tagger_eng')
```

## Example Text

```python
text = """
My name is Abdullah Hamdi. I am an AI engineer who is passionate about machine learning and natural language processing. 
I am currently building this NLP repository to strengthen my understanding of text processing and machine learning techniques. 
At the moment, I am working on this project while completing my military service in Salloum. 
Even though the environment is challenging and resources are limited, I believe that continuous learning is possible anywhere. 
My goal with this repository is to practice, experiment, and improve my skills step by step so that I can become a better AI engineer in the future.
"""
```

## Synonym Replacement

```python
aug = naw.SynonymAug(aug_src="wordnet")

augmented_text = aug.augment(text)

print(augmented_text)
```

## Random Word Swap

```python
aug = naw.RandomWordAug(action="swap")

aug.augment(text)
```

## Random Deletion

```python
aug = naw.RandomWordAug(action="delete")

aug.augment(text)
```

## Why Augmentation Matters
Text augmentation helps:

- increase training data
- reduce overfitting
- improve model robustness

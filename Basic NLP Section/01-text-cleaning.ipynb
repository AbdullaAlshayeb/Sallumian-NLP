# Text Cleaning

In this notebook we will learn the basic text cleaning steps used in Natural Language Processing (NLP).

Cleaning text is the first step before applying any NLP model because raw text often contains:
- punctuation
- special characters
- extra spaces
- inconsistent capitalization

## Import Libraries

```python
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords

nltk.download('stopwords')
```

## Example Dataset

```python

text = """
My name is Abdullah Hamdi. I am an AI engineer who is passionate about machine learning and natural language processing. 
I am currently building this NLP repository to strengthen my understanding of text processing and machine learning techniques. 
At the moment, I am working on this project while completing my military service in Salloum. 
Even though the environment is challenging and resources are limited, I believe that continuous learning is possible anywhere. 
My goal with this repository is to practice, experiment, and improve my skills step by step so that I can become a better AI engineer in the future.
"""

print(text)
```

## Convert Text to Lowercase

```python
text_lower = text.lower()

print(text_lower)
```

## Remove Punctuation

```python
text_no_punct = re.sub(r"[^\w\s]", "", text_lower)

print(text_no_punct)
```

## Remove Numbers (we don't have in our sample)

```python
text_no_numbers = re.sub(r"\d+", "", text_no_punct)

print(text_no_numbers)
```

## Remove Extra Spaces

```python
text_clean_spaces = re.sub(r"\s+", " ", text_no_punct).strip()

print(text_clean_spaces)
```

## Remove Stopwords

```python
stop_words = set(stopwords.words("english"))

words = text_clean_spaces.split()

filtered_words = [word for word in words if word not in stop_words]

clean_text = " ".join(filtered_words)

print(clean_text)
```

## Creating Cleaning Function

```python
def clean_text(text):
    
    text = text.lower() 
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    
    words = text.split()
    words = [word for word in words if word not in stop_words]
    
    return " ".join(words)
```

```python
clean_text(text)
```

## Compare Before and After

```python
print("Original Text: \n")
print(text)

print("Cleaned Text \n")
print(clean_text(text))
```

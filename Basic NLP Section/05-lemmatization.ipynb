# NLP Basics — Lemmatization

Lemmatization is the process of converting a word to its base or dictionary form, known as the lemma.

Unlike stemming, which simply removes suffixes, lemmatization uses linguistic knowledge and vocabulary to return meaningful base forms.

Examples:

running → run  
studies → study  
better → good  
am → be  

Lemmatization helps preserve the true meaning of words while reducing vocabulary size in Natural Language Processing tasks.

## Lemmatization with POS Tags

### Convert POS tags to WordNet format

```python
def get_wordnet_pos(word):
    tag = pos_tag([word])[0][1][0].upper()
    
    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV
    }

    return tag_dict.get(tag, wordnet.NOUN)
```

## Compare Stemming vs Lemmatization

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "studies", "better", "am", "playing"]

for word in words:
    stem = stemmer.stem(word)
    lemma = lemmatizer.lemmatize(word, get_wordnet_pos(word))
    
    print(word, "| Stem:", stem, "| Lemma:", lemma)
```

## When to Use Lemmatization

Lemmatization is preferred in tasks like:

search engines

chatbots

question answering

semantic analysis

because it preserves true word meaning.

## Key Takeaways

- Lemmatization converts words to their dictionary form.
- It is more accurate than stemming because it uses vocabulary and grammar rules.
- It often requires Part-of-Speech tagging to produce correct results.
- Lemmatization helps reduce vocabulary size while preserving word meaning.

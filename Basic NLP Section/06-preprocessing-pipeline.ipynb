# Text Preprocessing Pipeline

In the previous notebooks we explored several preprocessing techniques used in NLP:

- Text Cleaning
- Tokenization
- Stopword Removal
- Stemming
- Lemmatization

In real NLP systems, these steps are combined into a preprocessing pipeline that prepares raw text for machine learning models.

In this notebook we will build a reusable preprocessing pipeline that transforms raw text into cleaned and normalized tokens.

## Import Libraries

```python
import re
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

## Example Text

```python
text = """
My name is Abdullah Hamdi. I am an AI engineer who is passionate about machine learning and natural language processing.
I am currently building this NLP repository to strengthen my understanding of text processing and machine learning techniques.
At the moment, I am working on this project while completing my military service in Salloum.
Even though the environment is challenging and resources are limited, I believe that continuous learning is possible anywhere.
My goal with this repository is to practice, experiment, and improve my skills step by step so that I can become a better AI engineer in the future.
"""
```

## Initialize Tools

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

## The Functions 

### Cleaning Function

```python
def clean_text(text):

    text = text.lower()

    text = re.sub(r"[^\w\s]", "", text)

    text = re.sub(r"\d+", "", text)

    text = re.sub(r"\s+", " ", text).strip()

    return text
```

### Tokenization Function

```python
def tokenize_text(text):

    tokens = word_tokenize(text)

    return tokens
```

### Stopword Removal Function

```python
def remove_stopwords(tokens):

    filtered = [word for word in tokens if word not in stop_words]

    return filtered
```

### Lemmatization Function

```python
def lemmatize_tokens(tokens):

    lemmas = [lemmatizer.lemmatize(word) for word in tokens]

    return lemmas
```

## The Pipeline Function

```python
def preprocess(text):

    text = clean_text(text)

    tokens = tokenize_text(text)

    tokens = remove_stopwords(tokens)

    tokens = lemmatize_tokens(tokens)

    return tokens
```

```python
processed_text = preprocess(text)

print(processed_text[:30])
```

## Apply Pipeline to Multiple Texts

```python
texts = [
    "I love working with NLP.",
    "Machine learning models require clean text.",
    "Text preprocessing is an important step."
]

processed = [preprocess(t) for t in texts]

print(processed)
```

## Why Pipelines Matter

Preprocessing pipelines allow NLP systems to process large amounts of text consistently.

Instead of repeating preprocessing code multiple times, we define a reusable function that can be applied to any dataset.

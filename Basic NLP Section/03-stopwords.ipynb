# Stopword Removal

Stopwords are very common words that usually do not carry significant meaning in text analysis.

Examples of stopwords include:

the, is, am, are, and, in, on, at, a, an

Removing stopwords helps reduce noise in text data and improves the efficiency of many NLP models.

In this notebook we will learn:

- What stopwords are
- Why they are removed
- How to remove them using NLTK
- How to create custom stopwords

## Import Libraries

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download("stopwords")
nltk.download("punkt")
```

## Example Text

```python
text = """
My name is Abdullah Hamdi. I am an AI engineer who is passionate about machine learning and natural language processing.
I am currently building this NLP repository to strengthen my understanding of text processing and machine learning techniques.
At the moment, I am working on this project while completing my military service in Salloum.
Even though the environment is challenging and resources are limited, I believe that continuous learning is possible anywhere.
My goal with this repository is to practice, experiment, and improve my skills step by step so that I can become a better AI engineer in the future.
"""
```

## Tokenize the Text

```python
tokens = word_tokenize(text)

print(tokens)
```

## Load Stopwords

```python
stop_words = set(stopwords.words("english"))

print(stop_words)
```

## Remove Stopwords

```python
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print(filtered_tokens)
```

## Compare Before and After

```python
print("Original token count:", len(tokens))
print("After stopword removal:", len(filtered_tokens))
```

## Visualize Removed Words

```python
removed_words = [word for word in tokens if word.lower() in stop_words]

print(removed_words[:20])
```

## Key Takeaways

- Stopwords are very common words that add little meaning to text analysis.
- Removing stopwords reduces noise and improves efficiency.
- Stopword removal usually happens after tokenization.
- Stopwords should not always be removed depending on the NLP task.

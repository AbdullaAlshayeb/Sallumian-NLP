# Bag of Words

Bag of Words (BoW) is one of the simplest methods for converting text into numerical features.

The idea is to represent a document by counting how many times each word appears.

Steps:
1. Build a vocabulary of all unique words.
2. Count how many times each word appears in each document.
3. Represent the document as a vector of word counts.

## Import Libraries

```python
from sklearn.feature_extraction.text import CountVectorizer
```

## Example Dataset

```python
documents = [
    "I love natural language processing",
    "I love machine learning",
    "Machine learning is part of artificial intelligence",
    "Natural language processing is interesting"
]
```

## Initialize the Vectorizer

```python
vectorizer = CountVectorizer()
```

## Fit and Transform the Text

```python
X = vectorizer.fit_transform(documents)
```

## View Vocabulary

```python
print(vectorizer.vocabulary_)
```

## Convert to Matrix

```python
print(X.toarray())
```

## Display as DataFrame

```python
import pandas as pd

bow_df = pd.DataFrame(
    X.toarray(),
    columns=vectorizer.get_feature_names_out()
)

bow_df
```

## Limitations of Bag of Words

- Ignores word order
- Treats all words with equal importance
- Frequent words dominate the representation

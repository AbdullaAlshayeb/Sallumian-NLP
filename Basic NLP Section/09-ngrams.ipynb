# N-Grams

N-grams are sequences of N consecutive words in a text.

Examples using the sentence:

"I love natural language processing"

Unigram (1 word):
I, love, natural, language, processing

Bigram (2 words):
I love, love natural, natural language, language processing

Trigram (3 words):
I love natural, love natural language, natural language processing

N-grams help capture some word order and context in text representation.

## Import Library

```python
from sklearn.feature_extraction.text import CountVectorizer
```

## Example DataSet

```python
documents = [
    "I love natural language processing",
    "Natural language processing is powerful",
    "Machine learning is amazing"
]
```

## Unigrams

```python
vectorizer = CountVectorizer(ngram_range=(1,1))

X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
```

## Bigrams

```python
vectorizer = CountVectorizer(ngram_range=(2,2))

X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
```

## Unigram + Bigram

```python
vectorizer = CountVectorizer(ngram_range=(1,2))

X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
```

## Convert to DataFrame

```python
import pandas as pd

ngram_df = pd.DataFrame(
    X.toarray(),
    columns=vectorizer.get_feature_names_out()
)

ngram_df
```

## Key Idea
 N-grams allow models to capture local word relationships that Bag of Words alone cannot represent.

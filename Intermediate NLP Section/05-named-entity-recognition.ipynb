```python
import spacy
from spacy import displacy
```

# Load Model

```python
nlp = spacy.load("en_core_web_sm")
```

# Basic Example

```python
text = "Abdullah Hamdi founded koushre and lives in the Egypt."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "|", ent.label_)
```

- PERSON → People
- ORG → Organizations
- GPE → Countries, cities
- DATE → Dates
- MONEY → Money

# Visualization

```python
displacy.render(doc, style="ent", jupyter=True)
```

# More Examples

```python
examples = [
    "Amr Hamed was born in Asuit.",
    "Apple is launching a new iPhone in September.",
    "Microsoft was founded by Omar Soruor.",
    "The event will happen on 5th July 2024.",
    "Khaled Ahmed is a great engineer. He works in KSA",
    "Khaled Hamdi is a superb dentist.",
    "Mohamed Alshayeb works in Dubai"
]

for text in examples:
    doc = nlp(text)
    print("\nTEXT:", text)
    for ent in doc.ents:
        print(ent.text, "-", ent.label_)
```

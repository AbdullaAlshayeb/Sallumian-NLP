#  Concept Explanation

Neural networks (like in PyTorch) only understand **numbers**, not text.

So we must convert:

```
"I love NLP" → numbers
```

---

###  Old Method: Bag of Words (BoW)

* Sparse vectors
* No meaning
* No context

Example:

```
"king" ≠ "queen" (no relationship captured)
```

---

###  Solution: Embeddings

A **word embedding** is:

> A dense vector that represents a word’s meaning

Example:

```
"king"   → [0.25, -0.1, 0.8, ...]
"queen"  → [0.27, -0.12, 0.75, ...]
```

 Similar words → similar vectors

---

###  Key Idea

Embedding Layer = a **lookup table** that maps:

```
word_index → dense vector
```

---

#  Why It Matters

Without embeddings:

* Models treat words as **independent tokens**
* No semantic understanding

With embeddings:

* Capture meaning
* Capture similarity
* Enable deep NLP models

 This is the foundation of:

* LSTM
* Transformers
* BERT

---

#  Simple Example (Intuition)

Think of embeddings like:

 A **map of words in space**

* "cat" and "dog" → close together 
* "car" → far away 

So the model understands:

```
cat ≈ dog
cat ≠ car
```


# Import Libraries

````python
import torch
import torch.nn as nn
````

# Example

````python
sentences = [
    "i love nlp",
    "i love machine learning",
    "nlp is fun"
]
````

# Build Vocabulary

````python
words = set(" ".join(sentences).split())

word2idx = {word: i for i, word in enumerate(words)}

vocab_size = len(word2idx)
print(word2idx)
````

# Convert Sentences to Indices

````python
def encode(sentence):
    return [word2idx[word] for word in sentence.split()]

encoded_sentences = [encode(s) for s in sentences]
print(encoded_sentences)
````

## Padding

````python
from torch.nn.utils.rnn import pad_sequence

encoded_tensors = [torch.tensor(seq) for seq in encoded_sentences]
padded = pad_sequence(encoded_tensors, batch_first=True)

print(padded)
````

## Embedding Layer

````python
embedding_dim = 5

embedding = nn.Embedding(
    num_embeddings=vocab_size,
    embedding_dim=embedding_dim
)
````

## Apply Embedding

````python
embedded_output = embedding(padded)

print(embedded_output.shape)
````

# Visualization / Results

````python
print(embedding.weight)
````

## Compare Words

````python
word = "nlp"
idx = word2idx[word]

print(f"Embedding for '{word}':")
print(embedding.weight[idx])
````
